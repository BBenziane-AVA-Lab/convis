```python
%pylab inline
import convis
```

```python
stimulus = np.zeros((5000,50,50))
T,X,Y = np.meshgrid(np.arange(stimulus.shape[0]),
                    np.arange(stimulus.shape[1]),
                    np.arange(stimulus.shape[2]),indexing='ij')
stimulus += 0.5*(1.0+np.sin(0.1 * T + 0.5*X + 0.3*Y))
convis.describe(stimulus)
```

```python
convis.base.default_resolution.input_luminosity_range = 1.0
convis.base.default_resolution.pixel_per_degree = 5.0
convis.base.default_resolution.steps_per_second = 200.0
retina = convis.retina.Retina(filter_epsilon = 0.01, debug=True)
retina.add_output(retina.ganglion_input_layers[0])
retina.add_output(retina.ganglion_input_layers[1])
retina.add_output(retina.bipol)
retina.add_output(retina.opl)
print retina.draw_simple_diagram()
```

```python
retina.resolution.filter_epsilon = 0.01
retina.reset_parameters()
retina.clear_states()
%time o = retina.run_in_chunks(stimulus,500) # the first run compiles the function
```
